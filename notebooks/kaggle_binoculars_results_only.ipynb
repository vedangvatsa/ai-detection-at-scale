```python
import glob, os, shutil
# Keep only results CSVs in /kaggle/working
for f in glob.glob("/kaggle/working/*"):
    if os.path.isfile(f):
        os.remove(f)
    elif os.path.isdir(f):
        shutil.rmtree(f)
os.makedirs("/kaggle/working/results", exist_ok=True)
for f in glob.glob("results/*.csv"):
    shutil.copy(f, "/kaggle/working/results/")
    print("Copied:", f)
print("\nOutput ready for download.")
```
